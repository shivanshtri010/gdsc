```python
#extracting the list of column names from the DataFrame
c = con.columns.tolist()
#removing these columns because they are not useful for prediction
c.remove('ID')
c.remove('Segmentation')
c.remove('train_y_n')

new_name = 'Segmentation'
#printing the remaining useful column names
print(c)
```

```python
# creating dummy variables for the Gender column and changing them into binary values
con=pd.get_dummies(con,prefix='Gender',columns=['Gender'],drop_first=True)
```

```python
# filling the Not Available values with Yes
con['Ever_Married']=con['Ever_Married'].fillna('Yes')
```

```python
#creating dummy variables for the Ever married column and changing them into binary values
con=pd.get_dummies(con,prefix='Married',columns=['Ever_Married'],drop_first=True)
```

```python
#filling the Not Available values with Yes
con['Graduated']=con['Graduated'].fillna('Yes')
```

```python
#creating dummy variables for the Graduated column and changing them into binary values
con=pd.get_dummies(con,prefix='Graduated',columns=['Graduated'],drop_first=True)
```

```python
#filling the Not Available values with Yes
con['Profession']=con['Profession'].fillna('Yes')
```

```python
#converting Pofession into string data type
con['Profession']=con['Profession'].astype('str')
```

```python
#encoding into numeric variable
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
con['Prof']=le.fit_transform(con['Profession'])
```

```python
sns.countplot(con['Prof'],hue=con['Profession'])
```

```python
#dropping the 'Profession' column from the DataFrame 
con.drop('Profession',axis=1,inplace=True)
```

```python
#filling the Not Available values with Yes
con['Work_Experience'].fillna(con['Work_Experience'].mean(),inplace=True)
```

```python
#converting into numeric

# identify rows where 'Spending_Score' is equal to 'Low', 'Average', or 'High', and then assigns the corresponding integer values 1, 2, or 3 to those rows, respectively
con.loc[con['Spending_Score']=='Low','Spending_Score']=1
con.loc[con['Spending_Score']=='Average','Spending_Score']=2
con.loc[con['Spending_Score']=='High','Spending_Score']=3

# converting the data type into integer
con['Spending_Score']=con['Spending_Score'].astype('int')
```

```python
##filling the Not Available values with Yes
con['Family_Size'].fillna(round(con['Family_Size'].mean()),inplace=True)
```

```python
#filling the Not Available values with Yes
con['Var_1'].fillna('Cat_6',inplace=True)

#modifying only the last character of each string in the column
con['Var_1']=con['Var_1'].apply(lambda x:x[-1])

#converting the data type into integer
con['Var_1']=con['Var_1'].astype('int')
```

```python
#splitting the DataFrame into training and evaluation datasets 
from sklearn.model_selection import train_test_split
df_train, df_eval = train_test_split(con[con['train_y_n']==1], test_size=0.40, random_state=101, shuffle=True, stratify=con[con['train_y_n']==1][new_name])
```

```python
#transforming both training and evaluation datasets 
le = preprocessing.LabelEncoder()
df_train['Segmentation']=le.fit_transform(df_train['Segmentation'])
df_eval['Segmentation']=le.fit_transform(df_eval['Segmentation'])
```

```python
import lightgbm as lgb
from sklearn import preprocessing
from sklearn.metrics import mean_squared_log_error, mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import log_loss
from sklearn.metrics import classification_report
import seaborn as sns
from collections import Counter
sns.set_style('darkgrid')
```

```python
#defining the hyperparameters
params = {}
params['learning_rate'] = 0.04
params['max_depth'] = 18
params['n_estimators'] = 3000
params['objective'] = 'multiclass'
params['boosting_type'] = 'gbdt'
params['subsample'] = 0.7
params['random_state'] = 42
params['colsample_bytree']=0.7
params['min_data_in_leaf'] = 55
params['reg_alpha'] = 1.7
params['reg_lambda'] = 1.11
```

```python
#extracting the list of column names from the DataFrame
feature_cols = df_train.columns.tolist()

#removing these columns because they are not useful for prediction
feature_cols.remove('ID')
feature_cols.remove('Segmentation')
feature_cols.remove('train_y_n')
label_col = 'Segmentation'
print(feature_cols)
```

```python
#list of categorical columns
cat_cols=['Spending_Score','Family_Size','Var_1','Gender_Male','Married_Yes','Graduated_Yes','Prof']
```

```python
# creating a new LightGBM classifier model
clf = lgb.LGBMClassifier(**params)

#fitting the LightGBM classifier
clf.fit(df_train[feature_cols], df_train[label_col], early_stopping_rounds=100, eval_set=[(df_train[feature_cols], df_train[label_col]), (df_eval[feature_cols], df_eval[label_col])], eval_metric='multi_error', verbose=True, categorical_feature=cat_cols)

# computing the accuracy score 
eval_score = accuracy_score(df_eval[label_col], clf.predict(df_eval[feature_cols]))

print('Eval ACC: {}'.format(eval_score))
```

```python
#spliting the dataframe into train and test
train=con[con['train_y_n']==1]
test=con[con['train_y_n']==0]
```

```python
# performing a left merge on the ID column
sub=pd.merge(left=test['ID'],right=train[['ID','Segmentation']],how='left',on='ID')
```

```python
#creating the actual dataframe
actual_test=(test[test['ID'].isin(train['ID'])==False])
```

```python
# array of predicted labels 
pred=clf.predict(actual_test[feature_cols])
```

```python
#assigning the predicted labels to the Segmentation column of the actual_test
pred=le.inverse_transform(pred)
actual_test['Segmentation']=pred
```

```python
# creating a submission file by merging the predicted segments with the customer ID
l=actual_test[['ID','Segmentation']]
r=sub[sub['Segmentation'].isnull()==False]
fr=[l,r]
sub=pd.concat(fr)
```

```python
# storing the data generated in a CSV file
sub[['ID','Segmentation']].to_csv('predicted_output.csv',index = False)
```
